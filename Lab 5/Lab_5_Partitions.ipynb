## 1. Write a code that produces the Bell numbers that count the possible partitions of a set fo $n$ elements.

**Hints**:
-  Output the n-th Bell Number - [stackexchange](https://codegolf.stackexchange.com/questions/132379/output-the-n-th-bell-number).

- Bell number [wikipedia](https://en.wikipedia.org/wiki/Bell_number)






```python

def bellNumber(n):

    bell = [[0 for i in range(n+1)] for j in range(n+1)]
    bell[0][0] = 1
    for i in range(1, n+1):

        # Explicitly fill for j = 0
        bell[i][0] = bell[i-1][i-1]

        # Fill for remaining values of j
        for j in range(1, i+1):
            bell[i][j] = bell[i-1][j-1] + bell[i][j-1]

    return bell[n][0]

# Driver program
for n in range(6):
    print('Bell Number', n, 'is', bellNumber(n))

# This code is contributed by Soumen Ghosh at https://www.geeksforgeeks.org/bell-numbers-number-of-ways-to-partition-a-set/
```

## 2.Write a code that that produces the list of all partititions of the set of $[n]=\{ 1,2,3,4 \cdots , n-1, n\}$ with $0 \le n \le 10$.

```python
lista_sumas_iguales=[]

def print_partition(ans):
    """
    Function to print a partition
    """
    for i in ans:
        print("{", end=" ")
        for element in i:
            print(element, end=" ")
        print("}", end=" ")
    print()


def partition_set(set, index, ans):
    """
    Function to generate all partitions
    """
    if index == len(set):
        # If we have considered all elements in the set, print the partition
        global lista_sumas_iguales
        print_partition(ans)
        lista_s=[]
        lista_sumas=[]
        for lista in ans:
          lista_s.append(sum(lista))
          lista_sumas.append(lista)
        if len(lista_s)>1:
          if all(x == lista_s[0] for x in lista_s):
            for sublista in lista_sumas:
              lista_sumas_iguales.append(sublista.copy())
            print("En la particion anterior La suma de los elemetos es la misma.")

        return

    # For each subset in the partition, add the current element to it and recall
    for i in range(len(ans)):
        ans[i].append(set[index])
        partition_set(set, index + 1, ans)
        ans[i].pop()

    # Add the current element as a singleton subset and recall
    ans.append([set[index]])
    partition_set(set, index + 1, ans)
    ans.pop()


def all_partitions(set):
    """
    Function to generate all partitions for a given set
    """
    ans = []
    partition_set(set, 0, ans)


if __name__ == "__main__":
    # The size of the set
    n= int(input("Escribe un numero entre 1 y 9: "))

    # Initialize the set as {1, 2, ..., n}
    set = list(range(1, n + 1))

    print("All partitions of the set will be:")
    # Generate all partitions of the set
    all_partitions(set)

    print(f"las particiones con las cuales la suma de los elementos es la misma es: {lista_sumas_iguales}")
```

## 3. Write a code that produces the list of all partititions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 10$ and the sum of elements of the partition which is the a constant $\frac{n(n+1)}{2}$.

```python
from itertools import combinations

def partitions(n):
    if n == 0:
        yield []
        return
    for p in partitions(n-1):
        yield [1] + p
        if p and (len(p) < 2 or p[1] > p[0] + 1):
            yield [p[0] + 1] + p[1:]

def constant_sum_partitions(n):
    target_sum = n * (n + 1) // 2
    for partition in partitions(n):
        if sum(partition) == target_sum:
            yield partition

for n in range(11):
    print(f"Partitions of [n] with sum {n * (n + 1) // 2}:")
    for partition in constant_sum_partitions(n):
        print(partition)
    print()
```
