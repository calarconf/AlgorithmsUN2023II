## Fibonacci recursive algorithm $O\left( \left(  \frac{1+\sqrt(5)}{2}  \right)^n  lg\ n\right)$ [Relaxed asymptotic bound $O(2^n\ lg\ n)$]

```
# fibR - Fibonacci Recursive O( ( (1+sqrt(5)/2 )^n lg n ) [Relaxed asymptotic bound O(2^n lg n)]
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)    # for large nÂ´s  the addition will take lg n stetp because the numebers are very large
```

```
import time
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(46)
y = []
for n in x:
    start_time = time.time()
    fibR(n)
    y.append(time.time()-start_time)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Recursive Fibonacci')
plt.plot(x, x, label='y=n', linestyle='--')

# Adding labels, title, and legend
plt.xlabel('n')
plt.ylabel('Execution time (s)')
plt.title('Recursive Fibonacci Execution Time')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

# Fibonacci iterartive algoritm (forward) $O(n\ lgn)$

```
# fibI - Fibonacci Iterartive (forward) O(n lgn)
def fibI(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```
import time
import matplotlib.pyplot as plt

# Iterative Fibonacci function
def fibI(n):
    if n == 0:
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a + b
    return b

# List to store execution times
execution_times = []

# Range of n values
n_values = range(1001)

# Measure execution time for each n
for n in n_values:
    start_time = time.time()
    fibI(n)
    end_time = time.time()
    execution_times.append(end_time - start_time)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times, marker='o', label='Execution Time of fibI(n)')
plt.plot(n_values, [n  for n in n_values], label='y = n ', linestyle='--')  # Scaling y=n for better comparison

plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Iterative Fibonacci Function with y = n for Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

# Fibonacci memoization (dynamic programming)  algoritm  (backward) $O(n\ lgn)$

```
# fibI - Fibonacci Memoization (dynamic programming) (backward) O(n lg n)

def fibM(n):
    memo = {}
    def fib(n):
        if n==0 :
            return 0
        if n==1 :
            return 1
        if (n-2) not in memo:
            memo[n-2] = fib(n-2)
        if (n-1) not in memo:
            memo[n-1] = fib(n-1)
        return memo[n-1] + memo[n-2]
    return fib(n)
```

```
import time
import matplotlib.pyplot as plt

# Fibonacci function with memoization (dynamic programming)
def fibM(n):
    memo = {}
    def fib(n):
        if n == 0:
            return 0
        if n == 1:
            return 1
        if (n-2) not in memo:
            memo[n-2] = fib(n-2)
        if (n-1) not in memo:
            memo[n-1] = fib(n-1)
        return memo[n-1] + memo[n-2]
    return fib(n)

# List to store execution times
execution_times = []

# Range of n values
n_values = range(1001)

# Measure execution time for each n
for n in n_values:
    start_time = time.time()
    fibM(n)
    end_time = time.time()
    execution_times.append(end_time - start_time)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times, marker='o', label='Execution Time of fibM(n)')
plt.plot(n_values, [n  for n in n_values], label='y = n ', linestyle='--')  # Scaling y=n for better comparison

plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Memoized Fibonacci Function with y = n for Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
